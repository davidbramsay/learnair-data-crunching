```python
%matplotlib inline
import matplotlib 
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt  
import pandas as pd
from dateutil import parser
import datetime
import pytz

localtz = pytz.timezone('US/Eastern')

sns.set(style='white', palette='Set2')

df = pd.read_pickle('full_mlarray.pkl')

df.index = df.index.tz_convert('US/Eastern')
```

```python
#RMS fit - scale factor and offset
#NOT USEFUL - unrealistic and large offset, which doesn't match reality (a lot of ~0 wind).  Better not to minimize offset.

#find a,b where we minimize sqrt( sum((a*scaled_ws+b)-ws)^2 / n ), same as minimzing sum((a*scaled_ws+b)-ws)^2 

ref_col = 'ws'
scaled_col = 'scaled_arduino_ws'
new_col = 'lmse_scaled_arduino_ws'

import math
from scipy.optimize import minimize

total_rmse = math.sqrt( np.nansum((df[scaled_col].values - df[ref_col].values)**2) / len(df) )

def minimize_me(x):
    return np.nansum(((x[0]*df[scaled_col].values + x[1]) - df[ref_col].values)**2)

x0 = np.array([1.,0.])
res = minimize(minimize_me, x0, method='nelder-mead',
                options={'xtol': 1e-8, 'disp': True})
print('nelder-mead minimization= %s\n' % res.x)

res2 = minimize(minimize_me, x0, method='Powell',
                options={'disp': True})
print('powell minimization= %s\n' % res2.x)

res3 = minimize(minimize_me, x0, method='BFGS',
                options={'disp': True})
print('BFGS minimization= %s\n' % res3.x)

df[new_col] = df[scaled_col]*res.x[0]+res.x[1]


fig, ax = plt.subplots(figsize=(9, 9))
sns.regplot(scaled_col, ref_col, data=df, fit_reg=False)
plt.ylim(0,20)
plt.xlim(0,20)


total_rmse = math.sqrt( np.nansum((df[scaled_col].values - df[ref_col].values)**2) / len(df) )
print 'original rmse = %s' % total_rmse

fig, ax = plt.subplots(figsize=(9, 9))
sns.regplot(new_col, ref_col, data=df, fit_reg=False)
plt.ylim(0,20)
plt.xlim(0,20)

total_rmse = math.sqrt( np.nansum((df[new_col].values - df[ref_col].values)**2) / len(df) )
print 'after optimization rmse = %s' % total_rmse


df.drop('lmse_scaled_arduino_ws', axis=1, inplace=True)
```

```python
#-OR- RMS fit - *JUST* scale factor, no offset

#find a where we minimize sqrt( sum((a*scaled_ws)-ws)^2 / n ), same as minimzing sum((a*scaled_ws+b)-ws)^2 

ref_col = 'ws'
scaled_col = 'scaled_arduino_ws'
new_col = 'lmse_scaled_arduino_ws'

import math
from scipy.optimize import minimize

total_rmse = math.sqrt( np.nansum((df[scaled_col].values - df[ref_col].values)**2) / len(df) )

def minimize_me(x):
    return np.nansum(((x*df[scaled_col].values) - df[ref_col].values)**2)

x0 = 1.0
res = minimize(minimize_me, x0, method='nelder-mead',
                options={'xtol': 1e-8, 'disp': True})
print('nelder-mead minimization= %s\n' % res.x)

res2 = minimize(minimize_me, x0, method='Powell',
                options={'disp': True})
print('powell minimization= %s\n' % res2.x)

res3 = minimize(minimize_me, x0, method='BFGS',
                options={'disp': True})
print('BFGS minimization= %s\n' % res3.x)

df[new_col] = df[scaled_col]*res.x

xmin=0
xmax=20

fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(14,6))
plt.sca(ax1)
sns.regplot(ref_col, scaled_col, data=df, ax=ax1, fit_reg=False, scatter_kws={'linewidths':.1,'edgecolor':'k',"alpha":0.1,"s":50}, line_kws={"alpha":.8, "lw":4, 'c':'k'})
plt.plot([xmin,xmax],[xmin,xmax], ls='dashed', lw=4, color='k', alpha=.6)
plt.ylim(xmin,xmax)
plt.xlim(xmin,xmax)
plt.legend(['Ideal Relationship'], loc=2)
plt.ylabel('Pressure Measurement', fontsize=13)
plt.xlabel('EPA Windspeed Measurement (m/s)', fontsize=13)
plt.title('Before LMSE Minimization')

total_rmse = math.sqrt( np.nansum((df[scaled_col].values - df[ref_col].values)**2) / len(df) )
print 'original rmse = %s' % total_rmse

plt.sca(ax2)
sns.regplot(ref_col, new_col, data=df, ax=ax2, fit_reg=False, scatter_kws={'linewidths':.1,'edgecolor':'k',"alpha":0.1,"s":50}, line_kws={"alpha":.8, "lw":4, 'c':'k'})
plt.plot([xmin,xmax],[xmin,xmax], ls='dashed', lw=4, color='k', alpha=.6)
plt.ylim(xmin,xmax)
plt.xlim(xmin,xmax)
plt.legend(['Ideal Relationship'], loc=2)
plt.ylabel('Pressure Measurement with LMSE Correction', fontsize=13)
plt.xlabel('EPA Windspeed Measurement (m/s)', fontsize=13)
plt.title('After LMSE Minimization')

fig.suptitle('Pressure Sensor Wind Speed Comparison Before and After LMSE Correction', y=1.01, fontsize=15)

total_rmse = math.sqrt( np.nansum((df[new_col].values - df[ref_col].values)**2) / len(df) )
print 'after optimization rmse = %s' % total_rmse

plt.savefig('./figs/ws_compare_lmse_raw.png', bbox_inches='tight', dpi=200)
```

```python
def plot_2vals(val1, val2, legend, ylabel, title, savename, leg_loc=1, frameon=False):
    fig = plt.figure(figsize=(12, 9))  
    ax = plt.subplot(111)  
    fig.autofmt_xdate()
    ax.fmt_xdata = matplotlib.dates.DateFormatter('%m-%d')
    plt.plot_date(df.index.values, df[val1].values, tz=localtz, ls='solid', marker=None, lw=2.5)
    plt.plot(df.index.values, df[val2].values, lw=2.5, alpha=0.7)
    plt.legend(legend, fontsize=13, loc=leg_loc, frameon=frameon )
    plt.ylabel(ylabel, fontsize=12)
    plt.title(title, fontsize=15)
    plt.savefig('./figs/' + savename + '.png', bbox_inches='tight', dpi=200)

def plot_2vals_dates(val1, val2, starttime, endtime, legend, ylabel, title, savename, leg_loc=1, frameon=False):
    
    fig = plt.figure(figsize=(12, 9))  
    ax = plt.subplot(111)  
    fig.autofmt_xdate()
    ax.fmt_xdata = matplotlib.dates.DateFormatter('%m-%d')
    plt.plot_date(df.ix[starttime:endtime].index.values, df.ix[starttime:endtime][val1].values, tz=localtz, ls='solid', marker=None, lw=2.5)
    plt.plot(df.ix[starttime:endtime].index.values, df.ix[starttime:endtime][val2].values, lw=2.5, alpha=0.7)
    plt.legend(legend, fontsize=13, loc=leg_loc, frameon=frameon)
    plt.ylabel(ylabel, fontsize=12)
    plt.title(title, fontsize=15)
    plt.savefig('./figs/' + savename + '.png', bbox_inches='tight', dpi=200)
```

```python
starttime = datetime.datetime(2016,4,20)
endtime = datetime.datetime(2016,4,21)

#plot lmse arduino vs actual for full period
plot_2vals('lmse_scaled_arduino_ws', 'ws',
          ['LMSE Corrected Pressure Windspeed','EPA Windspeed'],
          'Windspeed (m/s)',
          'LearnAirV1 Windspeed Measurement, 04/15-06/13',
          'ws_lmse_vs_real', frameon=True)

plot_2vals_dates('lmse_scaled_arduino_ws', 'ws', starttime, endtime,
          ['LMSE Corrected Pressure Windspeed','EPA Windspeed'],
          'Windspeed (m/s)',
          'LearnAirV1 Windspeed Measurement, 04/20',
          'ws_lmse_vs_real_zoomed')



#plot lmse moving average arduino vs moving average actual for full period
plot_2vals('lmse_avg_30_scaled_arduino_ws', 'avg_30_ws',
          ['LMSE Corrected 30min Avg Pressure Windspeed','30min Avg EPA Windspeed'],
          'Windspeed (m/s)',
          '30min Avg LearnAirV1 Windspeed Measurement, 04/15-06/13',
          'ws_avg_30_lmse_vs_real', frameon=True)

plot_2vals_dates('lmse_avg_30_scaled_arduino_ws', 'avg_30_ws', starttime, endtime,
          ['LMSE Corrected 30min Avg Pressure Windspeed','30min Avg EPA Windspeed'],
          'Windspeed (m/s)',
          '30min Avg LearnAirV1 Windspeed Measurement, 04/20',
          'ws_avg_30_lmse_vs_real_zoomed')
```
