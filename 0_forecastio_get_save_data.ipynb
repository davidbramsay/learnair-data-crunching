```python
#actually calls the API for each day from time to final_day, appends all hourly data, and saves
#it to a file

time = parser.parse('04/15/16')
final_day = parser.parse('06/13/16')
#54 days of arduino data, 52 days of sck data

time_step = datetime.timedelta(days=1)

final_write = []

while time <= final_day:
    returned_vals = forecastio.load_forecast(api_key, roxbury_gps['lat'], roxbury_gps['lon'], time)

    for x in returned_vals.hourly().data:
        final_write.append(x.d)
        
    time += time_step

    
with open('weathervals.txt', 'w') as outfile:
    json.dump(final_write, outfile)
```

```python
#opens previously generated file, iterates through and prints timestamp of each saved value
import pytz
eastern = pytz.timezone('US/Eastern')

with open('weathervals.txt') as infile:
    weathervals = json.load(infile)

for x in weathervals:
    print datetime.datetime.fromtimestamp(x['time'], tz=eastern)
```
